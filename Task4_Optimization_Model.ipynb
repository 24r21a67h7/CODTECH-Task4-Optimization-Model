```python
from pulp import *
```

```python
model = LpProblem("Profit_Maximization", LpMaximize)
```

```python
A = LpVariable('Product_A', lowBound=0, cat='Integer')
B = LpVariable('Product_B', lowBound=0, cat='Integer')
```

```python
model += 50 * A + 40 * B
```

```python
model += 2 * A + 1 * B <= 100
model += 1 * A + 1 * B <= 80
```

```python
model.solve()
```

```python
print("Status:", LpStatus[model.status])
```

```python
print("Product A:", A.varValue)
print("Product B:", B.varValue)
```

```python
print("Maximum Profit:", value(model.objective))
```
